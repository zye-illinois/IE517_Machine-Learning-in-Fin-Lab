# Listing 2-8: Correlation between Classification Target and Real Attributes

# Listing 2-9: Pearson’s Correlation Calculation for Attributes 10 versus 14 and 10 versus 16

```python
from math import sqrt
import sys
#calculate correlations between real-valued attributes
cols=['Coupon', 'Issued Amount', 'Maturity At Issue months', 'LiquidityScore',
       'LIQ SCORE', 'n_trades', 'volume_trades', 'total_median_size',
       'total_mean_size', 'n_days_trade', 'days_diff_max',
       'percent_intra_dealer', 'percent_uncapped', 'bond_type',
       'Client_Trade_Percentage', 'weekly_mean_volume', 'weekly_median_volume',
       'weekly_max_volume', 'weekly_min_volume', 'weekly_mean_ntrades',
       'weekly_median_ntrades']
dataRow10 = data.loc[9,cols]
dataRow14 = data.loc[13,cols]
dataRow16 = data.loc[15,cols]
mean10 = 0.0; mean14 = 0.0; mean16 = 0.0
numElt = len(dataRow10)
for i in range(numElt):
    mean10 += dataRow10[i]/numElt
    mean14 += dataRow14[i]/numElt
    mean16 += dataRow16[i]/numElt
var10 = 0.0; var14 = 0.0; var16 = 0.0
for i in range(numElt):
    var10 += (dataRow10[i] - mean10) * (dataRow10[i] - mean10)/numElt
    var14 += (dataRow14[i] - mean14) * (dataRow14[i] - mean14)/numElt
    var16 += (dataRow16[i] - mean16) * (dataRow16[i] - mean16)/numElt
corr1014 = 0.0; corr1016 = 0.0
for i in range(numElt):
    corr1014 += (dataRow10[i] - mean10) * \
    (dataRow14[i] - mean14) / (sqrt(var10*var14) * numElt)
    corr1016 += (dataRow10[i] - mean10) * \
    (dataRow16[i] - mean16) / (sqrt(var10*var16) * numElt)
sys.stdout.write("Correlation between attribute 10 and 14 \n")
print(corr1014)
sys.stdout.write(" \n")
sys.stdout.write("Correlation between attribute 10 and 16  \n")
print(corr1016)
sys.stdout.write(" \n")
```
